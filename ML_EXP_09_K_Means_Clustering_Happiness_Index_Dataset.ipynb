# **K-MEANS CLUSTERING (2D dataset)**

Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
import matplotlib.pyplot as plt
```

Importing Dataset

```python
data1=pd.read_csv('/content/drive/MyDrive/DATASETS/happines index.csv')
data1
```

Data Preprocessing

```python
encoder=LabelEncoder()
data1['Country']=encoder.fit_transform(data1['Country'])
```

```python
scaler=MinMaxScaler()
normalized_data1=scaler.fit_transform(data1)
```

Clustering

```python
inertia_values=[]
silhouette_scores=[]
for k in range(2,11):
  k_means_model=KMeans(n_clusters=k)
  k_means_model.fit(normalized_data1)
  silhouette_score=metrics.silhouette_score(normalized_data1,k_means_model.labels_)
  silhouette_scores.append(silhouette_score)
  inertia=k_means_model.inertia_
  inertia_values.append(inertia)
  plt.scatter(data1['Country'], data1['Score'], c=k_means_model.labels_, cmap='rainbow', label="points")
  plt.title(f"No. of clusters : {k}")
  plt.show()
```

PLotting Silhoette Score

```python
plt.figure(figsize=(8, 6))
plt.plot(range(2,11), silhouette_scores, marker='o', label='Silhouette Score')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.legend()
plt.grid(True)
plt.show()
```

Elbow PLot

```python
plt.figure(figsize=(8, 6))
plt.plot(range(2,11), inertia_values, marker='o', label='Inertia')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method - Inertia vs. Number of Clusters')
plt.legend()
plt.grid(True)
plt.show()
```
