# **DECESION TREE**

**Importing libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Importing dataset**

```python
#importing dataset
data=pd.read_csv('/content/drive/MyDrive/DATASETS/winequality.csv')
#displaying dataset
data
```

**Data preprocessing**

Handling missing and duplicate values

```python
#dropping rows with missing values
data.dropna(inplace=True)
#dropping duplicate rows
data.drop_duplicates(inplace=True)
```

Encoding

```python
from sklearn.preprocessing import LabelEncoder
#creating object
encoder=LabelEncoder()
#performing encoding
data['type']=encoder.fit_transform(data['type'])
```

Handling highly correlated values

```python
#setting figure size
plt.figure(figsize=(15,15))
#finding correlation
correlation=data.corr()
#visualizing correlation
sns.heatmap(correlation,annot=True,cmap='coolwarm')
```

Standardization

```python
from sklearn.preprocessing import StandardScaler
#creating object
scaler=StandardScaler()
#performing standardization
scaler.fit_transform(data)
```

**Identifying dependent and independent values**

```python
X=data.drop(columns=['type'])
Y=data['type']
```

**Creating decision tree classifier**

model 1

```python
from sklearn.tree import DecisionTreeClassifier
#creating object with max depth 5
model1=DecisionTreeClassifier(max_depth=4)
#fitting dat
model1.fit(X,Y)

#importing libraries for visualization
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image

#creating dot data
dot_data=export_graphviz(model1, out_file=None, feature_names=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality'],
                         class_names=['white','red'], rounded=True, filled=True)

#creating graph
graph = pydotplus.graph_from_dot_data(dot_data)

#writting image
graph.write_png("decisioin_tree.png")

#creating image
Image(graph.create_png())
```
