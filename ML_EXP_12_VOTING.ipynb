# **VOTING (REGRESSION)**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data1=pd.read_csv('/content/drive/MyDrive/DATASETS/Boston Housing.csv')
```

```python
data1
```

```python
data1.drop_duplicates(inplace=True)
data1.dropna(inplace=True)
```

```python
x=data1.drop(columns=['PRICE'])
y=data1['PRICE']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#model1=LinearRegression()
#model2=xgb.XGBRegressor()
model1=GradientBoostingRegressor()
model2=SVR()
model3=RandomForestRegressor()
```

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
```

```python
prediction_1=model1.predict(x_test)
prediction_2=model2.predict(x_test)
prediction_3=model3.predict(x_test)
```

```python
print(f"Mean squared error for Gradient boosting regression : {mean_squared_error(y_test,prediction_1)}")
print(f"Mean squared error for support vector regression : {mean_squared_error(y_test,prediction_2)}")
print(f"Mean squared error for random forest regression : {mean_squared_error(y_test,prediction_3) }")
```

```python
final_prediction=(prediction_1+prediction_2+prediction_3)/3
```

```python
print(f"Mean squared error of ensembled model : {mean_squared_error(y_test,final_prediction)}")
```

# **VOTING (CLASSIFICATION)**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier,VotingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
```

```python
data2=pd.read_csv('/content/drive/MyDrive/DATASETS/Heart Attack.csv')
data2
```

```python
data2.drop_duplicates(inplace=True)
data2.dropna(inplace=True)
```

```python
encoder=LabelEncoder()
data2['class']=encoder.fit_transform(data2['class'])
```

```python
x=data2.drop(columns=['class'])
y=data2['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model_1=LogisticRegression()
model_2=xgb.XGBClassifier()
model_3=RandomForestClassifier()
```

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
```

```python
prediction_1=model1.predict(x_test)
prediction_2=model2.predict(x_test)
prediction_3=model3.predict(x_test)
```

```python
print(f"Log loss for logistic regression  : {log_loss(y_test,prediction_1)}")
print(f"Log loss for XG Boost : {log_loss(y_test,prediction_2)}")
print(f"Log loss for Random forest : {log_loss(y_test,prediction_3)}")
```

```python
final_model=VotingClassifier([('lr',model_1),('xgb',model_2),('rf',model_3)],voting='hard')
```

```python
final_model.fit(x_train,y_train)
```

```python
prediction_final=final_model.predict(x_test)
print(f"Log loss for ensembled model : {log_loss(y_test,prediction_final)}")
```
