```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('/content/drive/MyDrive/DATASETS/obesity_data.csv')
data
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
encoder=LabelEncoder()
data['Gender']=encoder.fit_transform(data['Gender'])
data['ObesityCategory']=encoder.fit_transform(data['ObesityCategory'])
```

```python
x=data.drop(columns=['ObesityCategory'])
y=data['ObesityCategory']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
tree=DecisionTreeClassifier(random_state=42)
bagging=BaggingClassifier(base_estimator=tree,n_estimators=30,random_state=42)
```

```python
bagging.fit(x_train,y_train)
```

```python
prediction=bagging.predict(x_test)
accuracy=accuracy_score(y_test,prediction)
print(f"Accuracy is {int((accuracy)*100)}%")
```
