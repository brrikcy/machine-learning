IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

IMPORTING DATASET

```python
data=pd.read_csv('/content/tip_modified.csv')
#reducing size of dataset by indexing
data=data[:50]
print(data)
```

DATA PREPROCESSING

```python
#dropping duplicate values
data.drop_duplicates(inplace=True)
#dropping rows with missing values
data.dropna(inplace=True)
```

```python
#performing normalization
data=data.drop(columns=['sex'])
scaler=MinMaxScaler()
scaler.fit_transform(data)
```

**LINEAR REGRESSION**

>INITIALIZING VALUES

```python
#dependent variable
Y=data['tip']

#independent variable
X=[data['total_bill'],data['encoded_sex']]


#slope1
slope1=0.1

#slope2
slope2=0.6

#intercept
incpt=0.3

#learning rate
lr=0.000001

#total number of iterations to be performed
iterations=500

#initializing a list to store calculated error values
error=[]
```

> APPLYING GRADIENT DESCEND ALGORITHM

```python
# for loop to perform specified number of iterations (value of iterations variable)
for itr in range(iterations):
    # initializing values for intermediate error,slope and intercept cost
    error_cost = 0
    slope1_cost = 0
    slope2_cost = 0
    incpt_cost = 0


    # loop for calculating cost function
    for i in range(50):

        y_predicted = slope1 * (X[0][i]) + slope2 * (X[1][i]) + incpt

        # cost function
        error_cost += (Y[i] - y_predicted) ** 2

        # loop for implementing gradient descent function for the calculated error
        for j in range(50):


            # partial derivative w.r.t Y
            partial_deri_intcpt = 2 * (Y[j] - (slope1 * (X[0][j])) - (slope2 * X[1][j]) - incpt)

            # partial derivative w.r.t X1
            partial_deri_slope1 = (-2 * slope1) * (Y[j] - (slope1 * (X[0][j])) - (slope2 * (X[1][j])) - incpt)

            # partial derivative w.r.t X2
            partial_deri_slope2 = (-2 * slope2) * (Y[j] - (slope1 * (X[0][j])) - (slope2 * (X[1][j])) - incpt)

            # accumulating values of partial derivative in each iteration
            incpt_cost += partial_deri_intcpt
            slope1_cost += partial_deri_slope1
            slope2_cost += partial_deri_slope2


        # updating values of slope and intercept
        incpt = incpt - (lr * incpt_cost)
        slope1 = slope1 - (lr * slope1_cost)
        slope2 = slope2 - (lr * slope2_cost)

    # printing error,slope and intercept
    print(itr, incpt, slope1, slope2)


    # appending corresponding error to error list
    error.append(error_cost)

# printing error
print(error)
```

```python
final_error=Y-y_predicted
print(final_error)
se=np.sum(final_error**2)
mse=se/50
print(f"Mean squared error is: {mse}")
print(f"Slope 1 : {slope1}")
print(f"Slope 2 : {slope2}")
print(f" Intercept : {incpt}")
```
